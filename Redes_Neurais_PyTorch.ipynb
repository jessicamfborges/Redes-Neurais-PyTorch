```python
import torch
import numpy as np
from torch import nn
```

```python
#Construindo uma REDE NEURAL modelo simples para representar uma reta
class LineNetwork(nn.Module):
  def __init__(self):
    super().__init__()

    #a minha variavel recebe uma entrada (usa o sequential para varias camadas)
    self.layers = nn.Sequential(
        nn.Linear(1,1)
    )
  #forward ele informa o dado que entra
  def forward(self, x):
    return self.layers(x)
```

```python
#O PyTorch tem uma classe chamada dataset para construi um dataset
from torch.utils.data import Dataset, DataLoader
import torch.distributions.uniform as urand
```

Abaixo uma representação de uma
reta, usando um dataset algebrico. É implementada uma função len para dizer quantos dados existe e getitem para dizer qual dado eu tenho (funcao, interval, nsamples(amostra))

```python
class AlgebraDataset(Dataset):
  def __init__(self, funcao, interval, nsamples):
    X = urand.Uniform(interval[0], interval[1]).sample([nsamples])
    self.data = [(x, f(x)) for x in X]

  #informar quantos dados eu tenho
  def __len__(self):
    return len(self.data)

  #informar o dado que eu quero acessar
  def __getitem__(self, index):
    return self.data[index]
```
